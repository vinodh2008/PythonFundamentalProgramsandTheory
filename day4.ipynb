```python
# bubble sort in python 
# i want thorey first 
# bubble sort is a simple sorting alogorithm that repeatedely checks adjacenet elements and swaps them if tehy are in
# wrong order
# this process is repeated until the array is sorted
def bubble_sort(arr):
    for i in range(len(arr)):
        for j in range(len(arr)-i-1):
            if arr[j]>arr[j+1]:
                arr[j],arr[j+1]=arr[j+1],arr[j]

arr=[3,5,7,2,4]
bubble_sort(arr)
print("array after bubble sort :",arr)
```
