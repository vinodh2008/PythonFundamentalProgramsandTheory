```python
# 52. Write a C program to take an array as input and print the elements that are not duplicated (i.e., elements that appear only once)
def uniue_elements(arr):
    freq={}
    for i in arr:
        if i in freq:
            freq[i]=freq[i]+1
        else:
            freq[i]=1
    for key,value in freq.items():
        if value==1:
            print(key,end=" ")
uniue_elements([1,2,3,4,3,2,1,5,6,7,8,7,6])
```

```python
# 46. Write a C program to calculate the average of numbers in the range 7 to 50 that are divisible by 3 but not divisible by 9.
def average_div3_not9(start,end):
    total=0
    count=0
    for i in range(start,end+1):
        if i%3==0 and i%9!=0:

          total+=i
          count+=1
    if count>0:
        average=total/count
        print("Average:",average)
    else:
        print("No numbers found")
average_div3_not9(7,50)
```
