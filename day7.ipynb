```python
# sum of n numbers using recursion
# how to build the logic in the ways of recursion
def sum_of_n_numbers(n):
    # base case
    if n==1:
        return 1
    else:
        res=n+sum_of_n_numbers(n-1)
        return res
    
num=int(input("Enter a number: "))
result=sum_of_n_numbers(num)
print("The sum of first",num,"numbers is:",result)
```

```python
# 2. Create a function to count vowels and consonants in a given string.  
def count_vowel_or_consonant(st):
    vowels="aeiou"
    v_c=0
    c_c=0
    for i in st:
        if i.isalpha():
            if i.lower() in vowels:
                v_c=v_c+1
            else:
                c_c=c_c+1
                
    print(f"count of the vowels in {st} is {v_c}")
    print(f"the count of the  consonant in {st} is {c_c}")
     
    
string=input("enter the  your string  to count the no of consonants and vowels")
count_vowel_or_consonant(string)

    
```

```python
# 3. Create a list of numbers and write code to filter only prime numbers from it.  
def filter_prime(lis):
    temp=[]
    for i in lis:
        count=0
        
        if i==1:
            continue
        for j in range(2,i//2):
            if i%j==0:
                # temp.append(j)
                count=1
                break
        if count==0:
            temp.append(i)
       
                
    return temp
        
# creation of numbers from using the range() function
list1=list(range(1,21))
print(filter_prime(list1))
```

```python
# Create a program to check if a string is a palindrome.
def is_palindrome(string):
    if len(string)<=2:
        return "maximum string lenght is greater than 2 charcters"
        
    rev=""
    for i in string:
        rev=i+rev
    if rev==string:
        return True
    else:
        return False
    
  
string=input("enter the string to check whether it is palindrome or not")
print(is_palindrome(string))
```

```python
# Create a menu-driven program to perform basic list operations: add, delete, search, display.
user_list=[]
def menu():
    print("Welcome to the list opertions Menu .Select the oper that you want  to perform")
    print("offering operations are:\n1.add element to list\n2.delete element from the list\n3.search element in the list\4.display the list elements\n5.eixt")
    while True:
        ch=int(input('\nenter the choice that you want to '))
        if ch==1:
            add()
            
        elif ch==2:
            delete()
        elif ch==3:
            search()
        elif ch==4:
            display()
        elif ch==5:
            exit()
        else:
            print("Invalid input")

def add():
    ele=input("enter the element")
    user_list.append(ele)
    print(f"{ele} is successfully inserted")
def delete():
    ele=input("enter the element to delete")
    user_list.remove(ele)
    print(f'{ele} is deleted successfully')
def search():
    is_search=False
    ele=input("enter the element to search")
    for i in range(len(user_list)):
        if ele ==user_list[i]:
            pos=i
            is_search=True
            break
    if is_search:
        print("element is found at index :",pos)
    else:
        print("element is not found")
def display():
    for i in user_list:
        print(i,end=" ")

    
if '__name__'=='__name__':
    menu()
```
