```python
class Rectange:
    width=" "
    height=" "
    def __init__(self,w,h):
        self.width=w
        self.height=h
    def area(self):
        res=self.width*self.height
        return res
        
    def perimeter(self):
        res=2*(self.height+self.width)
        return res  
    def display_info(self):
        print(f"width:{self.width} and height :{self.height}")
r1=Rectange(4,5)
r1.display_info()
print("Area of the r1 object is :",r1.area())
print("Perimeter of the r1 object",r1.perimeter())
print("-"*30)
#############
```

```python
# Write a class Employee with attributes name and salary. Add a method to increase salary by a percentage.
class Employee:
    name=" "
    salary=None
    def __init__(self,name,salary):
        self.name=name
        self.salary=salary
    def increase_salary(self,perc):
        
        increament=(perc/100)*self.salary
        self.salary=self.salary+increament
    def get_salary(self):
        print("current salary:",self.salary)
e1=Employee("vinodh",50000)
e1.increase_salary(5)
e1.get_salary()
    
        
    
    
```
