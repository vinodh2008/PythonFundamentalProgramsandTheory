```python
# creating the 3-d array
# you can also know the dimensions of by using the ndim()
one=nd.array([3,5,6,32])
two=nd.array([[3,5,6,32],[3,9,8,2]])
three=nd.array([[[3,5,6,32]],[[3,9,8,2]]])
print(nd.ndim(one))# it returns the  dimensions of specic array in numpy library
print(nd.ndim(two))

print(nd.ndim(three))
```

```python
# you can also create the array with zeros initialy that you can specify the no of dimensions 
# you can know the no dimensions that your array but dont know the values that you want to add after
# you can do it with zeros method it takes the no of arguments for no rows,coulms
zero=nd.zeros(6) # 1D array with zeroes
print(zero)
```
